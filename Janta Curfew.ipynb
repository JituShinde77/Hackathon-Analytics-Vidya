```python
from sklearn.model_selection import train_test_split

y = df.default_payment_next_month
df = df.drop(['ID', 'default_payment_next_month'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.20, random_state = 40, stratify = y)

print (X_train.shape)
print (X_test.shape)
print (y.shape)
```

```python
test_df = pd.read_csv('test_O6kKpvt.csv')
test_df.head()
```

```python
test_df = test_df.drop(['ID'], axis = 1)
test_df.head()
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier

tuned_parameters = {
    "n_estimators": [1, 5, 10, 50, 100, 200, 300, 400, 500, 600, 700, 800, 1000], "max_depth": [1, 5, 7, 8, 10, 12, 14, 16, 18, 20, 25, 30, 35, 40, 50, 75, 100]    
}

DT = RandomForestClassifier(criterion='gini', n_jobs= 6, class_weight='balanced')

model = RandomizedSearchCV(DT, tuned_parameters, scoring='roc_auc', cv= 5, verbose = 10)

model.fit(df, y)
```

```python
from scipy import stats
import xgboost as xgb

params = {
    "n_estimators": stats.randint(10, 300), 
    "max_depth": stats.randint(2, 10),    
    "learning_rate": stats.uniform(0.01, 1),
     "colsample_bytree": [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]
}        
    
xg = xgb.XGBClassifier(booster='gbtree', objective='binary:logistic', reg_lambda=0, reg_alpha= 1, scale_pos_weight=3.52)
model = RandomizedSearchCV(xg, params, scoring='roc_auc',  cv=5, n_jobs= 3, random_state=12, n_iter= 10, verbose=10)
model.fit(X_train, y_train)
```

```python
model.best_score_
```

```python
pred2 = model.predict(X_test)
testdf2 = model.predict(test_df)
```

```python
pred_val_y = 0.5 * pred1 + 0.5 * pred2 

thresholds = []
for thresh in np.arange(0.1, 1, 0.01):
    thresh = np.round(thresh, 2)
    res = roc_auc_score(y_test, (pred_val_y > thresh).astype(int))
    thresholds.append([thresh, res])
    print("F1 score at threshold {0} is {1}".format(thresh, res))
thresholds.sort(key=lambda x: x[1], reverse=True)
best_thresh = thresholds[0][0]
print("Best threshold: ", best_thresh)
```

```python
pred = ((testdf1 + testdf2) > best_thresh ).astype(int)
```
